```python
import tensorflow as tf
import tensorflow_hub as hub

url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
mobilenetv2 = hub.KerasLayer(url, input_shape=(224,224,3)) #Los dos primeros números son los pixeles de la imagen y el tercer número es la cantidad de categorías

#https://tfhub.dev/
```

```python
#Congelar el modelo descargado
mobilenetv2.trainable = False
```

```python
modelo = tf.keras.Sequential([
    mobilenetv2,
    tf.keras.layers.Dense(3, activation='softmax') #Número Dense tiene que ser el mismo que el número de carpetas
])

#https://keras.io/api/models/sequential/
```

```python
#Compilar como siempre
modelo.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

#https://keras.io/api/models/model_training_apis/
```

```python
#Entrenar el modelo
EPOCAS = 50  # Este número puede aumentar para hacerlo más preciso. Dependerá del tiempo de servidor y de la calidad de la predicción

historial = modelo.fit(
    data_gen_entrenamiento, epochs=EPOCAS, batch_size=32, #número 32 es  el número de lotes cada que se actualiza el gradiente, default 32
    validation_data=data_gen_pruebas
)

#https://keras.io/api/models/model_training_apis/
```

# Aquí hay que poner un cacho de código para ver en qué se está fijando la red. Hay que recordar el ejemplo de los perros y los lobos, para evitar que la red se fije en alguna característica que no se nos haya ocurrido y que esté en las imágenes.

```python
#Gráficas de precisión
#Revisar las especificaciones de history para poder ver más graficas
#https://keras.io/guides/training_with_built_in_methods/

acc = historial.history['accuracy']
val_acc = historial.history['val_accuracy']

loss = historial.history['loss']
val_loss = historial.history['val_loss']

rango_epocas = range(50)

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(rango_epocas, acc, label='Precisión Entrenamiento')
plt.plot(rango_epocas, val_acc, label='Precisión Pruebas')
plt.legend(loc='lower right')
plt.title('Precisión de entrenamiento y pruebas')

plt.subplot(1,2,2)
plt.plot(rango_epocas, loss, label='Pérdida de entrenamiento')
plt.plot(rango_epocas, val_loss, label='Pérdida de pruebas')
plt.legend(loc='upper right')
plt.title('Pérdida de entrenamiento y pruebas')
plt.show()
```
